## Importância da Previsão de Demanda no Varejo

No varejo, especialmente para produtos como vinho, é crucial prever a demanda com precisão, pois o vinho pode ter uma vida útil limitada e altos custos de armazenamento. Previsões incorretas podem levar a excessos ou faltas de estoque, resultando em custos de armazenamento ou oportunidades de vendas perdidas. Uma previsão de demanda precisa pode ajudar:

**- Competição**

  As empresas podem gerenciar seus estoques de forma eficiente, evitando excessos ou falta de produtos.

**- Melhorar os Processos da Cadeia de Suprimentos:**

  As empresas podem realizar reposições de estoque em tempo hábil para evitar gargalos.

**- Planejar Promoções de Vendas:**

  Quando as previsões indicam um aumento na demanda, ações de marketing direcionadas podem ser preparadas.

Aqui, como mencionado na parte sobre competição, é importante entender o contexto geográfico, mesmo que você opere online. ## Análise de Séries Temporais

A análise de séries temporais é um componente fundamental da análise de dados, que lida com o estudo de conjuntos de dados que registram informações sequenciais ao longo de intervalos de tempo. Esse tipo de análise permite identificar, compreender e prever padrões, tendências e variações cíclicas nos dados. A importância da análise de séries temporais se estende a diversas áreas de aplicação, desde previsões meteorológicas e análise de mercados financeiros até a previsão de tendências de consumo e a otimização de processos operacionais na indústria. Por meio da capacidade de prever eventos futuros com base em dados históricos, a análise de séries temporais permite que tomadores de decisão em empresas e organizações desenvolvam estratégias informadas e minimizem riscos. [Link](https://gpt5.blog/zeitreihenanalyse-time-series-analysis/)


## Para que serve a análise de séries temporais?

A análise de séries temporais serve a uma variedade de propósitos, incluindo:

- Previsão de eventos futuros: Identificação de padrões em dados históricos para prever tendências ou comportamentos futuros (por exemplo, previsão de preços de ações).

- Identificação de tendências: Ajuda os tomadores de decisão a tomar decisões informadas com base em tendências observadas (por exemplo, na economia).

- Compreensão de variações cíclicas: Análise de padrões repetitivos ao longo do tempo, como sazonalidade em dados de vendas.

Essa técnica é essencial para transformar dados históricos em insights acionáveis, permitindo uma melhor gestão e planejamento em diversos setores.

## Conceitos Básicos da Análise de Séries Temporais

Conceitos Básicos da Análise de Séries Temporais
Série Temporal: Uma sequência de pontos de dados que são medidos e registrados em ordem temporal consecutiva.

Para utilizar a análise de séries temporais de forma eficaz, é importante estar familiarizado com alguns conceitos básicos:

- **Ponto de Dados:** Um único valor ou observação na série temporal.

- **Timestamp (Carimbo de Tempo):**O momento específico em que um ponto de dados foi registrado.

- **Tendência:** A direção geral em que a série temporal se move ao longo de um período mais longo.

Esses conceitos são essenciais para entender e analisar dados de séries temporais.

## Entendendo a Estacionariedade na Análise de Séries Temporais Estacionário:

 Uma série temporal é considerada estacionária quando suas propriedades estatísticas (como média e variância) permanecem constantes ao longo do tempo.

A estacionariedade é um conceito fundamental na análise de séries temporais, pois muitos métodos de análise assumem que a série é estacionária ou exigem medidas para transformar uma série não estacionária em estacionária. Existem dois tipos principais de estacionariedade:

- **Estacionariedade Fraca** (ou em Sentido Amplo): Apenas a média e a variância da série temporal permanecem constantes ao longo do tempo.
Exemplo: A média e a variância das vendas mensais não mudam ao longo dos anos.

- **Estacionariedade Forte (ou Estrita):**Todas as propriedades estatísticas da série temporal (incluindo momentos de ordem superior) permanecem constantes ao longo do tempo. Exemplo: Além da média e variância, a distribuição completa dos dados não muda ao longo do tempo.

A estacionariedade é crucial porque simplifica a modelagem e a previsão de séries temporais. Quando uma série não é estacionária, técnicas como diferenciação ou transformações podem ser aplicadas para alcançar a estacionariedade, permitindo o uso de modelos como ARIMA ou SARIMA.

## Diferentes Tipos de Séries Temporais

Existem diversos tipos de séries temporais, que devem ser analisadas e interpretadas de maneiras específicas:

- **Tendência (Trend):** Uma série temporal que apresenta um aumento ou diminuição de longo prazo.
Exemplo: Crescimento constante nas vendas ao longo de vários anos.

- **Sazonalidade (Seasonal):** Uma série temporal com padrões regulares que se repetem em um intervalo de tempo fixo (por exemplo, anualmente, mensalmente ou semanalmente). Exemplo: Aumento nas vendas de sorvete durante o verão.

- **Cíclica (Cyclical):** Séries temporais com padrões que não se repetem em um intervalo de tempo fixo e estão frequentemente associados a condições econômicas ou ciclos de negócios.
Exemplo: Flutuações nas vendas de imóveis devido a mudanças nas taxas de juros.

- **Ruído (Noise):** Variações aleatórias nos dados que não podem ser explicadas por tendências, sazonalidade ou ciclos. Exemplo: Flutuações imprevisíveis nas vendas devido a eventos aleatórios.

- **Regressores (Regressors):**  Variáveis externas que podem influenciar a série temporal e são usadas para melhorar a precisão das previsões.
Exemplo: Feriados, promoções ou eventos especiais que impactam as vendas.

Esses componentes ajudam a decompor e entender melhor as séries temporais, permitindo análises mais precisas e previsões mais confiáveis.



## O que é Autocorrelação em uma Série Temporal?


Autocorrelação é a medida da correlação entre uma série temporal e uma versão deslocada dela mesma.

Em outras palavras, é a medida da semelhança entre os dados em diferentes momentos no tempo.

Por exemplo, se você correlacionar a temperatura de hoje com a temperatura de ontem, descobrirá que existe uma correlação positiva.


<p align=center>
<img src="https://mariofilho.com/img/guia_ts/4.png" width="70%"></p>

Em outras palavras:
Se a temperatura estava alta ontem, é provável que a temperatura hoje também esteja alta. Lembre-se de que correlação não implica causalidade: a temperatura de ontem não causou a temperatura de hoje, mas elas estão relacionadas.

No gráfico, você pode observar que a autocorrelação mais alta ocorre com o dia anterior (defasagem 1) e diminui em comparação com dias mais distantes.

Se os dados forem completamente aleatórios, ou seja, se um valor não tiver relação com o valor anterior, a autocorrelação será zero. Um exemplo de autocorrelação muito próxima de zero é a correlação entre os retornos de dias consecutivos do índice DAX.

<p align=center>
<img src="https://mariofilho.com/img/guia_ts/5.png" width="70%"></p>

Preste atenção ao eixo Y, que neste caso varia de -0,25 a 0,25 para representar melhor a autocorrelação. As barras pretas indicam se a autocorrelação passa no teste de significância estatística.
Em outras palavras: valores dentro desse intervalo indicam que a autocorrelação medida não difere estatisticamente de zero; estamos apenas vendo desvios de zero devido ao ruído. Em geral, o retorno de um dia não está relacionado ao retorno do dia anterior.

No caso de uma autocorrelação negativa, podemos pensar em uma série de exercícios físicos. Se você correr 10 km hoje, é provável que esteja mais cansado amanhã e não consiga correr a mesma distância. Quanto mais você corre em um dia (período anterior), menos consegue correr no dia seguinte (período atual).

### **2.2.9 Autocorrelation**

```python
# Plot the autocorrelation (ACF) of bike sharing demand
fig, ax = plt.subplots(figsize = (10,6))
plot_acf(df_sales_sum['sales'], lags = 100, ax = ax)
plt.show()
```

Das ist ein Autokorrelationsplot (ACF - Autocorrelation Function). Er zeigt, wie stark die aktuellen Werte in deiner Zeitreihe mit früheren Werten korrelieren.

PACF measures the correlation between a time series and its lag, after removing the effects of all shorter lags.

Purpose: To identify the direct relationship between an observation and its lag, without intermediate effects.

Main Difference between ACF and PACF:
- ACF: Measures overall correlation at each lag, including indirect effects.
- PACF: Measures direct correlation at each lag, excluding indirect effects.

**Was bedeutet diese Grafik?**
- Y-Achse: Die Korrelation (Werte zwischen -1 und 1).
- X-Achse: Die Lag-Zeit, also wie viele Zeitschritte zurück verglichen werden.
- Punkte mit vertikalen Linien: Die Korrelation für verschiedene Zeitabstände (Lags).
- Blau schattierter Bereich: Der Konfidenzbereich. Punkte außerhalb dieses Bereichs sind statistisch signifikant.

**Analyse Grafik:**

1- Hohe Anfangskorrelation: Der erste Punkt (Lag 0) ist immer 1, weil ein Wert mit sich selbst perfekt korreliert. Danach sieht man, dass die Autokorrelation langsam abnimmt.

2- Langsame Abnahme: Die Korrelation bleibt für viele Lags positiv und nimmt allmählich ab. Das zeigt, dass dein Zeitreihendatensatz eine starke zeitliche Abhängigkeit hat.

3- Saisonale Muster: Einige Peaks und Dips könnten auf saisonale Effekte hindeuten. Falls dein Datensatz tägliche Verkaufszahlen enthält, könnte es wöchentliche Muster geben.

4- Signifikante Autokorrelation: Mehrere Lags liegen außerhalb des Konfidenzintervalls, was bedeutet, dass frühere Werte immer noch einen Einfluss auf spätere Werte haben.

**Was kann man daraus ableiten?**

- Wenn die Korrelation langsam abnimmt, deutet das auf eine starke Trendkomponente hin.
- Falls Peaks in regelmäßigen Abständen auftreten, könnte das auf Saisonalität hinweisen.
- Falls du ein Vorhersagemodell baust, solltest du überlegen, saisonale oder autoregressive Komponenten (z. B. ARIMA oder Prophet mit saisonalen Effekten) einzubeziehen.

**Wie überprüft man, ob eine Zeitreihe stationär ist?**

Es gibt viele Möglichkeiten, dies zu überprüfen. Zum Beispiel:

**Grafische Betrachtung** - In vielen Fällen können Sie während der exploratorischen Analyse bereits visuell erkennen, ob eine Zeitreihe stationär ist oder nicht.

In diesem Projekt werden wir später einen spezifischen statistischen Test durchführen, um festzustellen, ob unsere Zeitreihe stationär ist oder nicht: den Augmented Dickey Fuller (ADF)-Test.

### ADF-Test

Der ADF-Test (Augmented Dickey-Fuller-Test) ist einer der am häufigsten verwendeten statistischen Tests für Zeitreihen. Seine Nullhypothese und Alternativhypothese lauten:

Die Nullhypothese ($H_O$ ) des Tests besagt, dass die Zeitreihe nicht stationär ist. Das bedeutet, dass sie eine gewisse Abhängigkeit von der Zeit aufweist.

Die Alternativhypothese ($H_1$) lehnt die Nullhypothese ab, d. h., sie besagt, dass die Zeitreihe stationär ist.

##**3.4 Schlussfolgerungen – Nicht-Stationäre Zeitreihe**

Alles sieht vielversprechend aus, aber wie können wir sicherstellen, dass unsere Vorhersagen tatsächlich zuverlässig sind?

Dafür verwenden wir die folgenden Metriken:

- Mean Absolute Percentage Error (MAPE) – Zeigt, wie stark unsere Vorhersagen prozentual von den tatsächlichen Werten abweichen.
- Mean Absolute Error (MAE) – Gibt den absoluten Fehler zwischen den vorhergesagten und den tatsächlichen Werten an.
Unten haben wir eine Funktion, um MAPE und MAE für unsere Zeitreihe zu berechnen.

```python
# Definieren der Funktion
def make_comparison_dataframe(historical, forecast):
  return forecast.set_index('ds')[['yhat', 'yhat_lower',
                                'yhat_upper']].join(historical.set_index('ds'))
cmp_df = make_comparison_dataframe(df_ts, forecast)

def calculate_forecast_errors(df_ts, prediction_size):

  df = df_ts.copy()
  df['e'] = df['y'] - df['yhat']
  df['p'] = 100 * (df['e'] / df['y'])
  predicted_part = df[-prediction_size:]

  error_mean = lambda error_name: np.mean(np.abs(predicted_part[error_name]))

  return{'MAPE': error_mean('p'), 'MAE': error_mean('e')}

#  MAPE und MAE
for err_name, err_value in calculate_forecast_errors(cmp_df, prediction_size).items():
  print(err_name, err_value)
```

Der mittlere absolute Fehler (MAE) misst den durchschnittlichen absoluten Unterschied zwischen den vorhergesagten und den tatsächlichen Werten. Beachten Sie, dass dieser mit 1417,89 relativ hoch ist.

Der MAPE misst hingegen den Prognosefehler in Prozent in Bezug auf die tatsächlichen Werte. Ein MAPE von weniger als 10 % gilt als Hinweis auf ein genaues Modell. In diesem Fall haben wir 3 % erreicht, was ein gutes Ergebnis für das Modell darstellt. Trotzdem müssen wir die stationäre Zeitreihe testen, um den MAE weiter zu reduzieren.

                𝑦′=log(𝑦)y ′ = log(y)

> Ein guter Anfang ist die Anwendung des $log$, um die Größenordnung der Werte der TS zu reduzieren. Anschließend subtrahieren wir den gleitenden Durchschnitt eines bestimmten Zeitraums vom $log$ der TS.

```python
# Aufbereitung der Daten
df_ts.index = pd.to_datetime(df_ts['ds'], format="%Y-%m-%d")
df_ts.drop('ds', axis=1, inplace=True)
df_ts.head()
```

## **4.5 Final Modell**

### 1.Datenaufbereitung und Aufteilung

```python
import optuna
import numpy as np
import pandas as pd
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics

# ---------------------------
# 1. Datenaufbereitung und Aufteilung
# ---------------------------
#
# Chronologische Sortierung
df_log = df_log.sort_values(by='ds').reset_index(drop=True)

# Aufteilung in 70% Training, 15% Validierung, 15% Test
n = len(df_log)
train_end = int(n * 0.7)
val_end = int(n * 0.85)

train_df = df_log.iloc[:train_end].copy()
val_df   = df_log.iloc[train_end:val_end].copy()
test_df  = df_log.iloc[val_end:].copy()

# Indizes zurücksetzen
train_df.reset_index(drop=True, inplace=True)
val_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)

print("Datenaufteilung:")
print("Trainingsdaten:", train_df.shape)
print("Validierungsdaten:", val_df.shape)
print("Testdaten:", test_df.shape)
```

### 2.Hyperparameter-Tuning und Trainingsdaten mit Cross-Validation

```python
# ---------------------------
# 2. Hyperparameter-Tuning mit Optuna (auf Trainingsdaten mit Cross-Validation)
# ---------------------------
def objective(trial):
    # Definition des Parameterraums
    params = {
        'changepoint_prior_scale': trial.suggest_float('changepoint_prior_scale', 0.001, 0.1, log=True),
        'seasonality_prior_scale': trial.suggest_float('seasonality_prior_scale', 1.0, 15.0),
        'holidays_prior_scale': trial.suggest_float('holidays_prior_scale', 1.0, 10.0),
        'seasonality_mode': trial.suggest_categorical('seasonality_mode', ['additive', 'multiplicative']),
        'n_changepoints': trial.suggest_int('n_changepoints', 5, 25)
    }

    # Prophet-Modell initialisieren und trainieren (nur auf Trainingsdaten)
    model = Prophet(yearly_seasonality=True, weekly_seasonality=True, **params)
    model.add_country_holidays(country_name='US')
    model.fit(train_df)

    # Cross-Validation durchführen:
    # initial: Zeitraum der ersten Trainingsphase,
    # period: Schrittweite, horizon: Forecast-Horizont (hier 30 Tage)
    df_cv = cross_validation(model, initial='665 days', period='15 days', horizon='30 days', parallel="processes")
    df_metrics = performance_metrics(df_cv, metrics=['mape', 'mae'])

    avg_mape = df_metrics['mape'].mean()
    avg_mae = df_metrics['mae'].mean()
    trial.set_user_attr('avg_mae', avg_mae)

    return avg_mape  # Ziel: Minimierung des MAPE

# Optuna-Studie erstellen und optimieren
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=30)

# Beste Hyperparameter extrahieren
#best_params = study.best_params
#print("\n Beste Hyperparameter:", best_params)
#print(" Bestes Validierungs-MAPE (CV auf Training):", study.best_value)
```

```python
# Beste Hyperparameter extrahieren
best_params = study.best_params
print("\n  Beste Hyperparameter:", best_params)
print(" Bestes Validierungs-MAPE (CV auf Training):", study.best_value)
```

### 3.Finales Modelltraining und Evaluierung auf Testdaten

```python
# ---------------------------
# 3. Finales Modelltraining und Evaluierung auf Testdaten
# ---------------------------
# Finales Training: Kombinieren von Trainings- und Validierungsdaten
final_train = pd.concat([train_df, val_df]).reset_index(drop=True)

final_model = Prophet(yearly_seasonality=True, weekly_seasonality=True, **best_params)
final_model.add_country_holidays(country_name='US')
final_model.fit(final_train)

# Prognose für den Testzeitraum:
# Berechnen der Anzahl Tage im Testdatensatz
n_test = len(test_df)
future_test = final_model.make_future_dataframe(periods=n_test, freq='D')
forecast_test = final_model.predict(future_test)

# Vorhersagen mit Testdaten zusammenführen
cmp_df_test = forecast_test.set_index('ds')[['yhat']].join(test_df.set_index('ds'))

# Funktion zur Fehlerberechnung (MAPE & MAE)
def calculate_forecast_errors(df):
    df = df.dropna(subset=['y'])
    df['e'] = df['y'] - df['yhat']
    df['p'] = 100 * df['e'] / df['y']
    return {'MAPE': np.mean(np.abs(df['p'])), 'MAE': np.mean(np.abs(df['e']))}

test_errors = calculate_forecast_errors(cmp_df_test)

print("\n Test-Ergebnisse:")
print(" Test MAPE:", round(test_errors['MAPE'],3),"%")
print(" Test MAE:", round(test_errors['MAE'],3))

```

**3. Vorhersagemodell-Genauigkeit:**

 - Die vorhergesagten Werte (rot) liegen sehr nah an den tatsächlichen Werten (blau), was auf ein gutes Modell hindeutet.

 - Allerdings gibt es einige stärkere Ausschläge in den echten Daten (z. B. einzelne extreme Einbrüche), die das Modell nicht perfekt erfasst.



**Schlussfolgerungen:**

Die Analyse zeigt, dass die Weinverkäufe deutliche saisonale Muster aufweisen, mit wiederkehrenden Spitzenwerten, die wahrscheinlich durch Feiertage, besondere Anlässe oder saisonale Nachfrage beeinflusst werden. Gleichzeitig gibt es einen langfristig positiven Wachstumstrend, der darauf hindeutet, dass der Markt für diese Produkte stabil wächst.

Das verwendete Vorhersagemodell liefert insgesamt präzise Prognosen, da die vorhergesagten Werte eng mit den tatsächlichen Verkaufszahlen übereinstimmen. Kleinere Abweichungen, insbesondere bei plötzlichen Einbrüchen oder Ausreißern, deuten jedoch darauf hin, dass außergewöhnliche Ereignisse oder externe Faktoren die Nachfrage kurzfristig beeinflussen können. Eine zusätzliche Berücksichtigung solcher Ereignisse in zukünftigen Modellen könnte die Prognosegenauigkeit weiter verbessern.

Insgesamt liefert die Analyse wertvolle Einblicke für eine gezieltere Lagerplanung und Verkaufsstrategie, um sowohl saisonale Nachfragespitzen als auch langfristige Wachstumspotenziale optimal zu nutzen.

# 5.0 Zentrale Erkenntnisse und Analyseergebnisse

**1. Saisonale Nachfrage-Trends**

 - Es gibt ein klares saisonales Muster mit einer steigenden Nachfrage in der ersten Jahreshälfte und einem Rückgang im zweiten Halbjahr, besonders im Januar.
 - Diese Saisonalität sollte in Lagerhaltungs- und Verkaufsstrategien berücksichtigt werden.

**2. Regionale Analyse**

 - Die Region Burgundy dominiert mit 32 % des Bestands, gefolgt von Bordeaux (21 %) und Rhone (6 %).
 - Diese Regionen sind nicht nur mengenmäßig führend, sondern tragen auch erheblich zum Gesamtumsatz bei.

**3. Weintypen und Jahrgänge**

 - Rotwein (60 %) dominiert den Markt, gefolgt von Weißwein (31,5 %) und Schaumwein (4,6 %).


**4. Preis- und Umsatzanalyse**

   - Die meisten Weine liegen preislich unter 250 USD, mit Ausnahme des Jahrgangs 2010 (ca. 600 USD Durchschnittspreis).
   - Die Jahrgänge 2002 und 2007 haben möglicherweise hohe Preise aufgrund begrenzter Verfügbarkeit.
   -   Die teuersten Weine stammen aus 2017, was mit einer hohen Nachfrage in diesem Jahr zusammenhängen könnte.

**5. Verkaufsvolumen vs. Umsatz**

   - Der meistverkaufte Wein ist auch der umsatzstärkste: **Domaine Ponsot Chapelle-Chambertin Grand Cru**.
   - Zwei weitere Weine sind sowohl in den Top 10 der Verkaufszahlen als auch in den Top 10 des Umsatzes vertreten:

        - **Domaine Ponsot Clos de La Roche Grand Cru Cuvée Vieilles Vignes** (zweithöchster Umsatz, siebthöchste Verkaufszahlen)
        - **Les Forts de Latour** (fünfthöchster Umsatz, zehnthöchste Verkaufszahlen)

  - Weine mit hohen Preisen generieren trotz geringerer Verkaufszahlen **erhebliche Umsätze.**


**6. Marktführer und strategische Bedeutung**

  - Die Marke Domaine Ponsot dominiert den Markt, sowohl im Verkaufsvolumen als auch im Umsatz.
  - Château Latour hat den zweithöchsten Umsatz, obwohl es nicht in den Top 10 der meistverkauften Weine ist.


##  **5.1 Empfehlungen:**

**1. Optimierung der Lagerhaltung:** Vorrangige Bevorratung von Weinen mit hoher Nachfrage und Umsatzstärke, insbesondere aus **Burgundy**, **Toskana** ,**Ribera del Duero** und **Bordeaux**.

 - Burgundy (Frankreich): Kleine Produktion, hohe Nachfrage – vor allem für
**PinotNoir und Chardonnay**.
 - Toskana (Italien): Weine wie Brunello di Montalcino oder Supertuscans haben oft eine stabile Wertentwicklung.
 - Auch die Regionen **Mendoza, Alentejo und Rioja** verzeichnen ein hohes Verkaufsvolumen, was auf das Interesse der Kunden an Weinen aus diesen Gebieten und diesem Weinstil hinweist.


**2. Gezielte Verkaufsaktionen:** Fokus auf umsatzstarke, aber weniger verkaufte Weine wie **Château Latour**.

  - Ein weiterer interessanter Punkt ist, dass Ribera del Duero mit nur fünf Produkten den 3. Platz beim Umsatz belegt.
  - Weitere Produkte von Domaine Ponsot anbieten, die das Unternehmen bisher noch nicht im Sortiment hat.
  - Mit mehr Kundenverkaufsdaten, mit Details über deren Geschmack und bevorzugte Stile, wäre es möglich, personalisierte Angebote zu erstellen und an der Segmentierung zu arbeiten.


**3. Anpassung an Saisonalität:** Steigerung der Bestände vor der Hauptnachfragezeit und gezielte Werbeaktionen gegen Jahresende.

  - Hochphase: Die erste Jahreshälfte, insbesondere die Monate **Mai bis Juli**, weisen die höchste Nachfrage auf.
  - Tiefphase: Die zweite Jahreshälfte, insbesondere **Januar**, zeigt einen deutlichen Rückgang der Nachfrage.


**4. Preisstrategien:** Berücksichtigung der Nachfrage-Preis-Dynamik, insbesondere für Jahrgänge mit hoher Wertentwicklung.

  - Wenn ein bestimmter Jahrgang teuer wird, weichen Kunden möglicherweise auf ähnliche, aber günstigere Alternativen aus. Eine Analyse von Substitutionsprodukten (z. B. Weine aus ähnlichen Regionen oder Jahrgängen) könnte hier wertvolle Erkenntnisse liefern.

  - Preis-Leistungs-Analyse: Falls das Ziel wäre, hochwertige Weine zu einem besseren Preis zu finden, könnten Regionen mit mittleren Preisen und guter Qualität (**z. B. Rioja oder Douro**) interessant sein.

Durch die Kombination von saisonalen Trends, regionalen Analysen und Preisstrategien können Einzelhändler fundierte Entscheidungen treffen, um sowohl die Rentabilität als auch die Kundenzufriedenheit zu maximieren.
